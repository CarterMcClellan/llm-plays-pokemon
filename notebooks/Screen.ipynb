```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Load and display the image
directory = "screenshots"
screenshots = os.listdir(directory)
screenshots = [os.path.join(directory, screenshot) for screenshot in screenshots]
img = Image.open(screenshots[0])
as_np = np.array(img)
```

```python
as_np.shape
```

```python
# Create figure and axis
fig, ax = plt.subplots()
ax.imshow(as_np)

x, y, width, height = 6, 103, 148, 33

# Create a Rectangle patch
rect = patches.Rectangle(
    (x, y),    # (x,y)
    width,               # width
    height,               # height
    linewidth=2,
    edgecolor='r',
    facecolor='none'
)

# Add the rectangle to the plot
ax.add_patch(rect)

plt.axis('off')  # Hide axes
plt.show()
```
